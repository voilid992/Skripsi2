# Convert File Xlsx to csv

```python
# Function to convert xlsx to csv
def convert_xlsx_to_csv(xlsx_path, csv_path):
    # Read the xlsx file
    xlsx_file = pd.read_excel(xlsx_path)
    
    # Write the xlsx file to a csv file
    xlsx_file.to_csv(csv_path, index=False)
```
